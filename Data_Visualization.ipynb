```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
plt.scatter(df['sepal_length'], df['petal_length'])
```

```python
x = [4,12,34,36,48,56,68,89]
y = [12,22,26,34,46,58,78,79]
```

```python
# scatter plot
plt.scatter(x, y)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.scatter(x, y, color = 'red')
```

```python
plt.figure(figsize=(16,9))
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.scatter(x, y, color = 'red', s=200)
```

```python
# Line plot
plt.plot(x, y)
```

```python
z = [56,51,50,47,41,23,22,18]
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, z, label = 'X vs. Z')
plt.plot(x, y, label = 'X vs. Y')
plt.legend()
```

```python
# Bar plot
studs = [45,67,123,178,88,120]
langs = ['C','C++','Java','Python','PHP','JavaScript']
```

```python
plt.bar(langs, studs, color = 'r')
```

```python
# Horizontal bar plot
plt.barh(langs, studs, color = 'r')
```

```python
# Pie chart
plt.pie(studs, labels=langs, autopct='%0.2f%%');
```

```python
# Histogram
p = [67,33,45,44,67,65,44,23,33,4,66,69,61,34,62,67,77]
plt.hist(p, bins=range(1,101,10));
```

```python
# Box plot
plt.boxplot(p);
```
